```python
Q1 = np.percentile(df['Culmen Depth (mm)'], 25,
interpolation = 'midpoint')

Q3 = np.percentile(df['Culmen Depth (mm)'], 75,
interpolation = 'midpoint')
IQR = Q3 - Q1

print("Old Shape: ", df.shape)

#The outlier base value is defined above and below datasets normal range namely Upper and Lower bounds, define the upper and the lower bound (1.5*IQR value is considered)
# Upper bound
upper = np.where(df['Culmen Depth (mm)'] >= (Q3+1.5*IQR))

# Lower bound
lower = np.where(df['Culmen Depth (mm)'] <= (Q1-1.5*IQR))

# Removing the Outliers
df.drop(upper[0], inplace = True)
df.drop(lower[0], inplace = True)

print("New Shape: ", df.shape)

sns.boxplot(x='Culmen Depth (mm)', data=df)
```
